```python
import polars as pl
pl.Config.set_tbl_rows(100)

filepath = "/data1/ccs_data/202603-rna-data/rna_data/20260520-rna/20260520_240601Y0004_Run0001_demuxed-metric/20260520_240601Y0004_Run0001_demuxed.gsmm2_aligned_metric_fact.csv"

data = pl.read_csv(filepath, separator="\t")

data = data.with_columns(
    pl.col("qname")
    .str.split("/")
    .list.last()                # 取最后一个元素
    .cast(pl.Int64)             # 转为整数
    .alias("sort_key")
).sort("sort_key").drop("sort_key")
```
